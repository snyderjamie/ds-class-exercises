```python
#Â Create a marvel_df pandas DataFrame with the given marvel data
marvel_data = [
    ['Spider-Man', 'male', 1962],
    ['Captain America', 'male', 1941],
    ['Wolverine', 'male', 1974],
    ['Iron Man', 'male', 1963],
    ['Thor', 'male', 1963],
    ['Thing', 'male', 1961],
    ['Mister Fantastic', 'male', 1961],
    ['Hulk', 'male', 1962],
    ['Beast', 'male', 1963],
    ['Invisible Woman', 'female', 1961],
    ['Storm', 'female', 1975],
    ['Namor', 'male', 1939],
    ['Hawkeye', 'male', 1964],
    ['Daredevil', 'male', 1964],
    ['Doctor Strange', 'male', 1963],
    ['Hank Pym', 'male', 1962],
    ['Scarlet Witch', 'female', 1964],
    ['Wasp', 'female', 1963],
    ['Black Widow', 'female', 1964],
    ['Vision', 'male', 1968]
]

marvel_df = pd.DataFrame.from_records(marvel_data)
marvel_df
```

```python
#Â Add column names to the marvel_df
col_names = ['name', 'sex', 'first_appearance']

marvel_df.columns = col_names
marvel_df
```

```python
#Â Add index names to the marvel_df (use the character name as index)

marvel_df.index = marvel_df['name']
marvel_df
```

```python
#Â Drop the name column as it's now the index

marvel_df = marvel_df.drop(columns='name')
# marvel_df = marvel_df.drop(['name'], axis=1)
marvel_df
```

```python
#Â Drop 'Namor' and 'Hank Pym' rows

marvel_df.drop(['Namor', 'Hank Pym'])
# marvel_df = marvel_df.drop(['Namor', 'Hank Pym'], axis=0)
```

## DataFrames manipulation and operations

```python
# Modify the first_appearance of 'Vision' to year 1964

marvel_df.at['Vision', 'first_appearance'] = 1964
# marvel_df.loc['Vision', 'first_appearance'] = 1964
marvel_df
```

```python
# Add a new column to marvel_df called 'years_since' with the years since first_appearance

marvel_df['years_since'] = 2019 - marvel_df['first_appearance']
marvel_df.head()
```

## DataFrames boolean arrays (also called masks)

```python
# Given the marvel_df pandas DataFrame, make a mask showing the female characters

marvel_df['sex'] == 'female'
```

```python
# Given the marvel_df pandas DataFrame, get the male characters

marvel_df[marvel_df['sex'] == 'male']
```

```python
# Given the marvel_df pandas DataFrame, get the characters with first_appearance after 1970

marvel_df[marvel_df['first_appearance'] > 1970]
```

```python
# Given the marvel_df pandas DataFrame, get the female characters with first_appearance after 1970

marvel_df[(marvel_df['sex'] == 'female') & (marvel_df['first_appearance'] > 1970)]
```

## DataFrames basic plottings

```python
# Reset the index names

marvel_df = marvel_df.reset_index()
```

```python
marvel_df
```

```python
#Â Plot the values of first_appearance

marvel_df.first_appearance.plot()
```

```python
#Â Plot a histogram (plot.hist) with values of first_appearance

fig, axes = plt.subplots(figsize=(12, 6))
plt.hist(marvel_df.first_appearance, bins=50, alpha=0.8,
          histtype='bar', color='powderblue',
          edgecolor='steelblue')
plt.title('First Appearance of Marvel Characters', fontsize=20)
plt.xlabel('Year')
plt.ylabel('Number of Characters')
```
